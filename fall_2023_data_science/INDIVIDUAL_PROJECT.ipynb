```python
# Your code here
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# List of tickers for the mentioned banks
selected_tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']

# Set the start and end dates
start = "2006-01-01"
end = "2023-11-20"

# Fetch data for the mentioned banks using yfinance
selected_banks_data = {}
for ticker in selected_tickers:
    selected_banks_data[ticker] = yf.download(ticker, start=start, end=end)

# Combine data for selected banks into a single DataFrame
selected_banks_df = pd.concat(selected_banks_data.values(), keys=selected_banks_data.keys(), axis=1)

# Plotting Close column for each selected bank in selected_banks_df
plt.figure(figsize=(12, 6))

for ticker in selected_tickers:
    plt.plot(selected_banks_df[ticker]['Close'], label=ticker)

plt.title('Closing Prices of Selected Banks')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

# Moving Average

Plot the 30-day moving averages for JPM CHASE and Morgan Stanley stocks for the first 600 days. Use will need to use the individual daframe that you have create for each stock in Part I.

```python
# Your code here

```

```python
import matplotlib.pyplot as plt
# Calculate the 30-day moving average for JPMorgan Chase
JPM['30_MA'] = JPM['Close'].rolling(window=30).mean()

# Calculate the 30-day moving average for Morgan Stanley
MS['30_MA'] = MS['Close'].rolling(window=30).mean()

# Plotting the Close prices and 30-day moving averages for JPMorgan Chase and Morgan Stanley
plt.figure(figsize=(10, 6))

# Plotting Close prices and 30-day moving averages together
plt.plot(JPM['Close'][:600], label='JPMorgan Chase Close', color='blue')
plt.plot(MS['Close'][:600], label='Morgan Stanley Close', color='green')
plt.plot(JPM['30_MA'][:600], label='JPMorgan Chase 30-day MA', linestyle='--', color='red')
plt.plot(MS['30_MA'][:600], label='Morgan Stanley 30-day MA', linestyle='--', color='orange')

plt.title('Closing Prices and 30-day Moving Averages for JPMorgan Chase and Morgan Stanley')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
plt.show()

```
